```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
# Coeficientes de Condutividade Térmica
#ca = 0.58
cr = 1.05

# Calor Específico
cea = 4180
cer = 1070

# Tempratura Ambiente
Tar = 293.15

# Coeficiente Convectivo do Ar
ccar = 10

# Dimensões do Recipiente

er = 0.005 # espessura

hr_ext = 0.09 # altura externa
hr_int = hr_ext - er # altura interna

rr_ext = 0.04 # raio externo
rr_int = rr_ext - er # raio interno

abr_ext = math.pi * rr_ext**2 # área da base externa
abr_int = math.pi * rr_int**2 # área da base interna

alr_ext = (2 * math.pi * rr_ext)*hr_ext # área lateral externa
alr_int = (2 * math.pi * rr_int)*hr_int # área lateral interna

ar_ext = alr_ext + abr_ext # área externa
ar_int = alr_int + abr_int # área interna
ar_meio = (ar_int + ar_ext)/2 # área intermediária

# Densidades
dr = 2403
da = 1000

# Volumes
va = abr_int * hr_int
vr = (abr_ext * hr_ext) - va

# Massas
mr = vr * dr
ma = va * da

# Resistências
Rconv_1 = 1/(ccar * abr_int)
Rcond_1 = (er/2)/(cr * ar_int)
Rcond_2 = (er/2)/(cr * ar_meio)
Rconv_2 = 1/(ccar * ar_ext)
```

```python
def modelo(x,t):
    Ta, Tr = x
    
    Q1 = (Ta - Tar)/Rconv_1
    Q2 = (Ta - Tr)/Rcond_1
    Q3 = (Tr - Tar)/(Rcond_2 + Rconv_2)
    
    dUadt = - Q1 - Q2
    dUrdt = Q2 - Q3
    
    dTadt = dUadt/(ma*cea)
    dTrdt = dUrdt/(mr*cer)
    
    dxdt = [dTadt,dTrdt]
    return dxdt
```

```python
# Lista Tempo
dt = 1e-2
t_max = 300
lista_t = np.arange(0, t_max, dt)

# Temperaturas Iniciais em K
T0_a = 353.15
T0_r = 298.15
lista_T0 = [T0_a, T0_r]

dTdt = odeint(modelo, lista_T0, lista_t)

dTadt = dTdt[:,0]-273.15
dTrdt = dTdt[:,1]-273.15

plt.plot(lista_t,dTadt,'r',label='Temperatura Água')
plt.plot(lista_t,dTrdt,'b',label='Temperatura Xícara')
plt.title('Temperatura da Água e da Xícara')
plt.xlabel('Tempo(s)')
plt.ylabel('Temperatura(Cº)')
plt.legend()
plt.grid()
plt.show()
```

```python
dt = 1e-2
t_max = 400
lista_t = np.arange(0, t_max, dt)

T0_a = 353.15
T0_r = 298.15
lista_T0 = [T0_a, T0_r]

lista_Ta = np.arange(353.15, 358.15, 0.5)

lista_tempos = []

for T0_a in lista_Ta:
    lista_T0 = [T0_a, T0_r]
    dTdt = odeint(modelo, lista_T0, lista_t)

    dTadt = dTdt[:,0]-273.15
 
    plt.plot(lista_t,dTadt)
    
plt.title('Temperatura da Água 80ºC - 85ºC')
plt.xlabel('Tempo(s)')
plt.ylabel('Temperatura(Cº)')
plt.grid()
plt.show()
```

```python
dt = 1e-2
t_max = 400
lista_t = np.arange(0, t_max, dt)

T0_a = 353.15
T0_r = 298.15
lista_T0 = [T0_a, T0_r]

lista_Ta = np.arange(353.15, 358.15, 0.5)

lista_tempos = []

for T0_a in lista_Ta:
    lista_T0 = [T0_a, T0_r]
    dTdt = odeint(modelo, lista_T0, lista_t)

    dTadt = dTdt[:,0]-273.15
    
    
    for i,T in enumerate(dTadt):
        if T <= 70:
            lista_tempos.append(lista_t[i])
            break
 
print(lista_tempos)
plt.plot(lista_Ta-273.15,lista_tempos,'b')
plt.plot(lista_Ta-273.15,lista_tempos,'ro',label='Temperatura Água')
plt.title('Tempo para a água chegar a 70ºC com diferentes temperaturas iniciais')
plt.ylabel('Tempo(s)')
plt.xlabel('Temperatura(Cº)')
plt.legend()
plt.grid()
plt.show()
```
